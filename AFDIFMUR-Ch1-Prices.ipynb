```python
amzn5 = web.DataReader("AMZN", 'yahoo', start, end)[['Close']]
amzn5['Diff'] = amzn5['Close'].diff()
amzn5['Up'] = amzn5['Diff'].apply(lambda x: 1 if x >0 else 0)
amzn5['Down'] = amzn5['Diff'].apply(lambda x: 1 if x <0 else 0)
amzn5 = amzn5[1:]
amzn5.head()
```

```python
amzn5['Up.Value'] = amzn5['Up']*amzn5['Diff']
amzn5['Down.Value'] = -amzn5['Down']*amzn5['Diff']
amzn5.head()
```

```python
amzn5['14D-Up-Mean'] = amzn5['Up.Value'].rolling(window=14).mean()
amzn5['14D-Down-Mean'] = amzn5['Down.Value'].rolling(window=14).mean()
amzn5.head(16)
```

```python
amzn5['14D-Up-ExpAvg'] = amzn5['Up.Value'].rolling(window=14).mean()
amzn5['14D-Up-ExpAvg'] = pd.ewma(amzn5['14D-Up-ExpAvg'], 14)
amzn5['14D-Down-ExpAvg'] = amzn5['Down.Value'].rolling(window=14).mean()
amzn5['14D-Down-ExpAvg'] = pd.ewma(amzn5['14D-Down-ExpAvg'], 14)
# amzn5['14D-Up-ExpAvg'] = amzn5['Up.Value'].rolling(window=14).mean()
# amzn5['14D-Up-ExpAvg'][14:] = (amzn5['14D-Up-ExpAvg'][14:].shift(-1)*13+amzn5['14D-Up-Mean'][14:])/14
# amzn5['14D-Down-ExpAvg'] = amzn5['Down.Value'].rolling(window=14).mean()
# amzn5['14D-Down-ExpAvg'][15:] = (amzn5['14D-Down-ExpAvg'][15:].shift(-1)*13+amzn5['Down.Value'][15:])/14
amzn5.head(20)
```

```python
amzn5['RS'] = amzn5['14D-Up-ExpAvg'] / amzn5['14D-Down-ExpAvg']
amzn5['RSI'] = 100-(100/(1+amzn5['RS']))
amzn5[14:20]
```
