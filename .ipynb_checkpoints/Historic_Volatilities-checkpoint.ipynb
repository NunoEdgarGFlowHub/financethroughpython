Analysis is based on this FT article, as inspiration: https://www.ft.com/content/fe4c064c-c50e-11e7-a1d2-6786f39ef675

```python
import pandas_datareader.data as web
import datetime
import pandas as pd
start = datetime.datetime(1925, 6, 1)
end = datetime.datetime.today()
import matplotlib.pyplot as plt
%matplotlib inline
gspc = web.DataReader("^GSPC", 'yahoo', start, end)[['Adj Close']]
gspc.plot(figsize=(20,10), grid=True)
```

```python
gspc['Returns'] = gspc['Adj Close'].pct_change()
gspc.head()
```

```python
gspc['AnnualRollingAverage'] = gspc['Returns'].rolling(window=252).mean()*100
gspc['AnnualRollingVar'] = gspc['Returns'].rolling(window=252).var()*100
gspc.tail(10)
```

```python
gspc['AnnualRollingAverage'].plot(figsize=(20,10), grid=True)
```

```python
gspc['AnnualRollingVar'].plot(figsize=(20,10), grid=True)
```

Although the values do not align with the FT's chart, the general pattern in the charts is more or less identical with similar high and low periods.
