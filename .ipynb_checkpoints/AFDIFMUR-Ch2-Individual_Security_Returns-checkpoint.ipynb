```python
import pandas_datareader.data as web
import pandas as pd
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
start = datetime.datetime(2010, 12, 31)
end = datetime.datetime(2013, 12, 31)
```

```python
ibm = web.DataReader("IBM", 'yahoo', start, end)
ibm.columns = ['IBM.Open','IBM.High','IBM.Low', 'IBM.Close','IBM.Adjusted','IBM.Volume']
ibm.head()
```

```python
ibm['IBM.pct.ret'] = ibm['IBM.Close'].pct_change()
ibmclose = ibm[['IBM.Close','IBM.pct.ret']]
ibmclose.head()
```

```python
ibmclose = ibmclose[1:]
ibmclose.head()
```

The tables above are similar to the output on page 57.

```python
ibm['IBM.tot.ret'] = ibm['IBM.Adjusted'].pct_change()
ibmadjusted = ibm[['IBM.Adjusted','IBM.tot.ret']]
ibmadjusted.head()
```

```python
ibmadjusted = ibmadjusted[1:]
ibmadjusted.head()
```

Similarly, the tables above are similar to the output on page 60-61.

```python
import numpy as np
ibm['IBM.log.ret'] = np.log(ibm['IBM.Adjusted']) - np.log(ibm['IBM.Adjusted'].shift(1))
ibmlog = ibm[['IBM.Close','IBM.log.ret']]
ibmlog = ibmlog[1:]
ibmlog.head()
```

Similarly, the tables above are similar to the output on page 62.

```python
ibmreturns = ibm[['IBM.pct.ret','IBM.tot.ret']].fillna(0)
ibmreturns['IBM.GrossDailyReturn'] = 1 + ibmreturns['IBM.tot.ret']
ibmreturns.head()
```

Similarly, the tables above are similar to the output on page 64.

```python
ibmreturns['IBM.GrossCumReturn'] = ibmreturns['IBM.GrossDailyReturn'].cumprod()
ibmreturns.head()
```

```python
ibmreturns['IBM.NetCumReturn'] = ibmreturns['IBM.GrossCumReturn'] - 1
ibmreturns.head()
```

Similarly, the tables above are similar to the output on page 65.

*** Same calculations using logarithmic returns have been skipped ***

```python
ibmreturns['IBM.GrossDailyReturn-PriceOnly'] = 1 + ibmreturns['IBM.pct.ret']
ibmreturns['IBM.GrossCumReturn-PriceOnly'] = ibmreturns['IBM.GrossDailyReturn-PriceOnly'].cumprod()
ibmreturns['IBM.NetCumReturn-PriceOnly'] = ibmreturns['IBM.GrossCumReturn-PriceOnly'] - 1
ibmreturns[['IBM.GrossCumReturn','IBM.GrossCumReturn-PriceOnly']].plot(figsize=(10,10), grid=True, style=['-','--'])
```

The chart above is similar to the one in Figure 2.1 on page 69.

Weekly and Monthly returns described on pages 68 to 73 can be calculated using .resample() method as described in the documentation here: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.resample.html

```python
ibmg = web.DataReader("IBM", 'yahoo', start, end)[['Close']]
ibmg.columns = ['IBM.Close']
ibmg['IBM.DailyReturn'] = ibmg['IBM.Close'].pct_change()+1
ibmg['IBM.CumReturn'] = ibmg['IBM.DailyReturn'].cumprod()
amzng = web.DataReader("AMZN", 'yahoo', start, end)[['Close']]
amzng.columns = ['AMZN.Close']
amzng['AMZN.DailyReturn'] = amzng['AMZN.Close'].pct_change()+1
amzng['AMZN.CumReturn'] = amzng['AMZN.DailyReturn'].cumprod()
gspcg = web.DataReader("^GSPC", 'yahoo', start, end)[['Close']]
gspcg.columns = ['GSPC.Close']
gspcg['GSPC.DailyReturn'] = gspcg['GSPC.Close'].pct_change()+1
gspcg['GSPC.CumReturn'] = gspcg['GSPC.DailyReturn'].cumprod()
```

```python
groupreturns = pd.concat([ibmg,amzng,gspcg], axis=1, join='outer',)
groupreturns[['AMZN.CumReturn','IBM.CumReturn','GSPC.CumReturn']].plot(figsize=(10,10), grid=True).axhline(1)
```

The chart above is similar to the one in Figure 2.2 on page 78.
