```python
start = datetime.datetime(1981, 12, 30)
end = datetime.datetime(2017, 12, 30)
DGS3MO = web.DataReader("DGS3MO", "fred", start, end)
DGS6MO = web.DataReader("DGS6MO", "fred", start, end)
DGS1 = web.DataReader("DGS1", "fred", start, end)
DGS2 = web.DataReader("DGS2", "fred", start, end)
DGS3 = web.DataReader("DGS3", "fred", start, end)
DGS5 = web.DataReader("DGS5", "fred", start, end)
DGS7 = web.DataReader("DGS7", "fred", start, end)
DGS10 = web.DataReader("DGS10", "fred", start, end)
DGS20 = web.DataReader("DGS20", "fred", start, end)
DGS30 = web.DataReader("DGS30", "fred", start, end)
tresy = pd.concat([DGS3MO, DGS6MO, DGS1, DGS2, DGS3, DGS5, DGS7, DGS10, DGS20, DGS30], join='outer', axis = 1)
```

```python
tresy = tresy['1990':'2017']
tresy.head()
```

```python
SPAB = web.DataReader("SPAB", "yahoo", start, end)['Adj Close']
JNK = web.DataReader("JNK", "yahoo", start, end)['Adj Close']
etfreturns = pd.concat([JNK, SPAB], join='outer', axis = 1)
etfreturns.columns = ['JNK','SPAB']
```

```python
etfreturns.plot(figsize=(20,10))
```

```python
etfreturns['JNK_Ret'] = etfreturns['JNK'].pct_change()+1
etfreturns['JNK_Val'] = etfreturns['JNK_Ret'].cumprod()

etfreturns['SPAB_Ret'] = etfreturns['SPAB'].pct_change()+1
etfreturns['SPAB_Val'] = etfreturns['SPAB_Ret'].cumprod()
```

```python
etfreturns[['JNK_Val','SPAB_Val']].plot(figsize=(20,10), grid=True)
```

Plot above is similar to the one on figure 8.14 on page 288 (ignoring additional necessary styling).
