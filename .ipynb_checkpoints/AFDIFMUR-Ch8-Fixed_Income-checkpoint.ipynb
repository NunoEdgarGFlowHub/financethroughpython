```python
start = datetime.datetime(1981, 12, 30)
end = datetime.datetime(2017, 12, 30)
DGS3MO = web.DataReader("DGS3MO", "fred", start, end)
DGS6MO = web.DataReader("DGS6MO", "fred", start, end)
DGS1 = web.DataReader("DGS1", "fred", start, end)
DGS2 = web.DataReader("DGS2", "fred", start, end)
DGS3 = web.DataReader("DGS3", "fred", start, end)
DGS5 = web.DataReader("DGS5", "fred", start, end)
DGS7 = web.DataReader("DGS7", "fred", start, end)
DGS10 = web.DataReader("DGS10", "fred", start, end)
DGS20 = web.DataReader("DGS20", "fred", start, end)
DGS30 = web.DataReader("DGS30", "fred", start, end)
tresy = pd.concat([DGS3MO, DGS6MO, DGS1, DGS2, DGS3, DGS5, DGS7, DGS10, DGS20, DGS30], join='outer', axis = 1)
```

```python
tresy = tresy['1990':'2017']
tresy.head()
```

```python
tresy.columns = [0.25, 0.5, 1, 2, 3, 5, 7, 10, 20, 30]
fig, ax = plt.subplots(figsize=(20,10))
tresy['2013-12-31':'2013-12-31'].T.plot(ax=ax,label='2013-12-31')
tresy['2000-11-24':'2000-11-24'].T.plot(ax=ax,label='2000-11-24')
tresy['2006-08-25':'2006-08-25'].T.plot(ax=ax,label='2006-08-25')
tresy['2010-01-11':'2010-01-11'].T.plot(ax=ax,label='2010-01-11')
```
