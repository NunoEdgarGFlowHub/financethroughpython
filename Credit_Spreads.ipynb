# IG vs HY vs 10y Treasury

```python
import pandas as pd
import pandas_datareader.data as web
import datetime
import matplotlib as plt
%matplotlib inline
start = datetime.datetime(1997, 1, 1)
end = datetime.datetime(2017, 11, 3)
usig = web.DataReader("BAMLC0A0CMSYTW", "fred", start, end)
ushy = web.DataReader("BAMLH0A0HYM2SYTW", "fred", start, end)
us10y = web.DataReader("DGS10", "fred", start, end)
```

```python
bigtable = pd.concat([usig, ushy,us10y], axis=1)
bigtable.columns = ['US IG', 'US HY', 'US10YT']
bigtable['USIG-USHY Spread'] = bigtable['US HY'] - bigtable['US IG']
bigtable.tail()
```

```python
bigtable.plot(figsize=(20,10),style=['-','-','-','--'])
```

# BBB vs BB vs CCC (All US)

```python
usbbb = web.DataReader("BAMLC0A4CBBBSYTW", "fred", start, end)
usbb = web.DataReader("BAMLH0A1HYBBSYTW", "fred", start, end)
usccc = web.DataReader("BAMLH0A3HYCSYTW", "fred", start, end)
bigtable2 = pd.concat([usbbb, usbb,usccc], axis=1)
bigtable2.columns = ['US-BBB', 'US-BB', 'US-CCC']
bigtable2['US BBB/BB Spread'] = bigtable2['US-BB'] - bigtable2['US-BBB']
bigtable2['US BB/CCC Spread'] = bigtable2['US-CCC'] - bigtable2['US-BB']
bigtable2.tail()
```

```python
bigtable2.plot(figsize=(20,10),style=['-','-','-','--','--'])
```
