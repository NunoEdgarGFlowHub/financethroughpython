```python
import pandas as pd
import pandas_datareader.data as web
from pandas.plotting import scatter_matrix
import datetime
import matplotlib.pyplot as plt
%matplotlib inline
start = datetime.datetime(2000, 6, 1)
end = datetime.datetime(2017, 10, 29)
brentoil = web.DataReader("DCOILBRENTEU", "fred", start, end)
brentoil.head()
```

```python
bp = web.DataReader("BP.L", 'yahoo', start, end)
bp = bp[['Adj Close']]
bp.head()
```

```python
shell = web.DataReader("RDSB.L", 'yahoo', start, end)
shell = shell[['Adj Close']]
shell.head()
```

```python
brentoil.plot(figsize=(20,10), grid=True)
plt.axvspan(pd.to_datetime('2008-04-01'), pd.to_datetime('2009-07-01'), color='grey', alpha=0.3, lw=0, label =  "Recession")
plt.axvspan(pd.to_datetime('2010-04-10'), pd.to_datetime('2010-07-15'), color='black', alpha=0.5, lw=0, label =  "Deepwater Horizon")
```

```python
table = brentoil.merge(bp, how='outer', left_index=True, right_index=True)
table = table.merge(shell, how='outer', left_index=True, right_index=True)
table.columns = ['Brent','BP','Shell']
table['Brent-R'] = table['Brent'].pct_change()
table['Brent-CR'] = (1 + table['Brent-R']).cumprod()*100
table['BP-R'] = table['BP'].pct_change()
table['BP-CR'] = (1 + table['BP-R']).cumprod()*100
table['Shell-R'] = table['Shell'].pct_change()
table['Shell-CR'] = (1 + table['Shell-R']).cumprod()*100
table.info()
```

```python
table[['Brent-CR','BP-CR','Shell-CR']].plot(figsize=(20,10), grid=True, title='Relative Cumulative Performance of Brent, Shell and BP')
plt.axvspan(pd.to_datetime('2008-04-01'), pd.to_datetime('2009-07-01'), color='grey', alpha=0.3, lw=0, label =  "Recession")
plt.axvspan(pd.to_datetime('2010-04-10'), pd.to_datetime('2010-07-15'), color='black', alpha=0.5, lw=0, label =  "Deepwater Horizon")
```

```python
returnstable = table[['Brent-R','BP-R','Shell-R']].dropna()
scatter_matrix(returnstable, alpha=0.6, diagonal='kde',figsize=(20,20), grid=True)
```

```python
returnstable.corr()
```

```python
# returnstable = table[['Brent-R','BP-R','Shell-R']]
Brent_BP_260corr = returnstable['Brent-R'].rolling(window=260).corr(returnstable['BP-R'])
Brent_Shell_260corr = returnstable['Brent-R'].rolling(window=260).corr(returnstable['Shell-R'])
BP_Shell_260corr = returnstable['BP-R'].rolling(window=260).corr(returnstable['Shell-R'])
Brent_BP_130corr = returnstable['Brent-R'].rolling(window=130).corr(returnstable['BP-R'])
Brent_Shell_130corr = returnstable['Brent-R'].rolling(window=130).corr(returnstable['Shell-R'])
BP_Shell_130corr = returnstable['BP-R'].rolling(window=130).corr(returnstable['Shell-R'])
```

```python
fig, ax = plt.subplots(figsize=(20,10))
Brent_BP_260corr.plot(ax=ax, label='Brent_BP_260c')
Brent_Shell_260corr.plot(ax=ax, label='Brent_Shell_260c')
BP_Shell_260corr.plot(ax=ax, label='BP_Shell_260c')
plt.axvspan(pd.to_datetime('2008-04-01'), pd.to_datetime('2009-07-01'), color='grey', alpha=0.3, lw=0, label =  "Recession")
plt.axvspan(pd.to_datetime('2010-04-10'), pd.to_datetime('2010-07-15'), color='black', alpha=0.5, lw=0, label =  "Deepwater Horizon")
plt.legend()
plt.show()
```
