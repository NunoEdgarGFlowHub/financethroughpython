# Historic Correlation between FTSE100 and GBPEUR

```python
ftse_gbp_corr = bigtable_returns[['^FTSE','GBPEUR=X']].dropna()
ftse_gbp_corr['^FTSE'].dropna().rolling(window=100).corr(ftse_gbp_corr['GBPEUR=X']).plot(figsize=(20,10))
```

```python
fig, axes = plt.subplots(20,figsize=(20,50),sharex=True, sharey=True)
i = 10
z=0
while i < 210:
    listtoplot = ftse_gbp_corr['^FTSE'].rolling(window=i).corr(ftse_gbp_corr['GBPEUR=X'])
    axes[z].plot(listtoplot.index,listtoplot.values)
    axes[z].set_title('Correlation Days: '+str(i))
    i = i + 10
    z+=1
plt.legend()
plt.show()
```
