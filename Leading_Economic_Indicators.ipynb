```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
```

## New Car Registration

```python
# Source: https://www.smmt.co.uk/vehicle-data/car-registrations/
cardata = pd.read_csv('smmtcarreg.csv', index_col=0, names=['Month','Count'])
cardata.index = pd.to_datetime(cardata.index,format='%b-%y')
cardata['CarsReg-Y-on-Ym'] = cardata['Count'].pct_change(periods = 12)*100
cardata.tail()
```

```python
# Source: https://www.ons.gov.uk/economy/grossdomesticproductgdp/timeseries/ihyr/pgdp
gdpdata = pd.read_csv('gdpdata.csv', index_col=0, names=['Date','GDP-Y-on-Yq'])
gdpdata.index = pd.to_datetime(gdpdata.index,format='%d/%m/%y')
pd.to_numeric(gdpdata['GDP-Y-on-Yq'])
gdpdata.tail()
```

```python
fig, ax = plt.subplots(figsize=(20,10))
cardata['CarsReg-Y-on-Ym'].plot(ax=ax, linestyle='--', marker='o')
gdpdata['GDP-Y-on-Yq'].plot(ax=ax, linestyle='-', marker='o')
plt.show()
```

```python
bigtable = cardata.merge(gdpdata, how='outer', left_index=True, right_index=True)
bigtable = bigtable[['Count','GDP-Y-on-Yq']][:-1]
bigtable = bigtable.resample('Q').sum()
bigtable['CarsReg-Y-on-Yq'] = bigtable['Count'].pct_change(periods = 4)*100
bigtable.tail()
```

```python
bigtable[['CarsReg-Y-on-Yq','GDP-Y-on-Yq']].plot(figsize=(20,10), grid=True, marker='o')
```
